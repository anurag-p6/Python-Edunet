```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


import warnings
warnings.filterwarnings('ignore')
```

```python
data = np.array([[85,90], [70, 75], [60, 65], [40, 50], [30, 35]])
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(data)
labels = kmeans.labels_
labels
```

```python
print("Cluster Center:\n\n", kmeans.cluster_centers_)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:, 1], color='red', marker='X')
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.title("K-Means Clustering of Students")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

colors = ['purple', 'green']
cluster_colors = np.array([colors[label] for label in labels])

plt.scatter(data[:, 0], data[:, 1], c = cluster_colors, alpha=0.6)


plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            color='red', marker='X', s=200, edgecolors='black', label='Centroids')

for i in range(len(kmeans.cluster_centers_)):
      cluster_center = kmeans.cluster_centers_[i]
      cluster_points = data[labels == 1]
      radius = np.max(np.linalg.norm(cluster_points - cluster_center, axis=1))
      circle = plt.Circle(cluster_center, radius, color='yellow', fill=False, linestyle='dashed',linewidth=2)
      plt.gca().add_patch(circle)

plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.title("K-Means Clustering of students")
plt.legend()
plt.show()
          
```
